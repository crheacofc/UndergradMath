## Euler's Method:
Euler's method for calculating the values of initial value differential equations works by computationally following the curved vector field of any given differential equation. Computationally, the method is very simple: We simply recaluculate the value by stepping along the slope of the solution. Most of my code snipets here on out will be in function form and I will only include the most relevant items (i.e. I will not include the error analysis part).

## Improved Euler methods

My Improved Euler method is similar to the original euler method

## Trapezoid Method/Midpoint Method

I used the trapezoid and midpoint methods since they are 4th order solvers and thus give better approximatoins.

## Runge-Kutta Method

I also included the classical Runge-Kutta fourth order method for some variety. (Note that the code below is for vectors)

```python
def g(x,y):
    return  -2*x-y
def l(t,y):
    return 2-e**(-4*t)-2*y
def k(x,y):
    return (y - x)/(y + x)
g_init = -1
ga=0
gb=0.5
k_init=1.
ka=0.0
kb=1.0
y_initl = 1
a = 0.0
b = 1.0
n= 10
ToL = 10**(-2)
```

```python
y = euler(g,g_init,ga,gb,n,ToL)
print("The best estimate according to the euler method for the differential equation y' = -2x-y is %s "%(y))
y1 = improved_euler(g,g_init,ga,gb,n,ToL)
print("The best estimate according to the improved euler method for the differential equation y' = -2x-y is %s "%(y1))
y2 = richardson_extrapolation(g,g_init,ga,gb,n,ToL)
print("The best estimate according to the richardson extrapolation of the euler method for the differential equation y' = -2x-y is %s "%(y2))
y3 = trapezium(g,g_init,ga,gb,n,ToL)
print("The best estimate according to the trapezium method for the differential equation y' = -2x-y is %s "%(y3))
y4 = Runge_Kutta(g,g_init,ga,gb,n,ToL)
print("The best estimate according to the Runge-Kutta method for the differential equation y' = -2x-y is %s "%(y4))

```

```python
y = euler(k,k_init,ka,kb,n,ToL)
print("The best estimate according to the euler method for the differential equation y' = 2-e^(-4t)-2y is %s "%(y))
y1 = improved_euler(k,k_init,ka,kb,n,ToL)
print("The best estimate according to the improved euler method for the differential equation y' = 2-e^(-4t)-2y is %s "%(y1))
y2 = richardson_extrapolation(k,k_init,ka,kb,n,ToL)
print("The best estimate according to the richardson extrapolation of the euler method for the differential equation y' = 2-e^(-4t)-2y is %s "%(y2))
y3 = trapezium(k,k_init,ka,kb,n,ToL)
print("The best estimate according to the trapezium method for the differential equation y' = 2-e^(-4t)-2y is %s "%(y3))
y4 = Runge_Kutta(k,k_init,ka,kb,n,ToL)
print("The best estimate according to the Runge-Kutta method for the differential equation y' = 2-e^(-4t)-2y is %s "%(y4))
```

```python
y = euler(l,y_initl,a,b,n,ToL)
print("The best estimate according to the euler method for the differential equation y' = (y-x)/(y+x) is %s "%(y))
y1 = improved_euler(l,y_initl,a,b,n,ToL)
print("The best estimate according to the improved euler method for the differential equation y' = (y-x)/(y+x) is %s "%(y1))
y2 = richardson_extrapolation(l,y_initl,a,b,n,ToL)
print("The best estimate according to the richardson extrapolation of the euler method for the differential equation y' = (y-x)/(y+x) is %s "%(y2))
y3 = trapezium(l,y_initl,ka,kb,n,ToL)
print("The best estimate according to the trapezium method for the differential equation y' = (y-x)/(y+x) is %s "%(y3))
y4 = Runge_Kutta(l,y_initl,a,b,n,ToL)
print("The best estimate according to the Runge-Kutta method for the differential equation y' = (y-x)/(y+x) is %s "%(y4))
```

### Vector ODE : Systems of differential equations!

Here I will demonstrate a few different minor examples of solving systems of differential equations.

```python
def xdot(t,y):
    z = np.matrix(np.zeros((1,2)))
    z[0,0] = y[:,1]**2-2*y[:,0]
    z[0,1] = y[:,0]-y[:,1]-t*y[:,1]**2
    return z
y = trapezium_vector(xdot,[0,1],0,1,100,2)
print(y[-1,0],y[-1,1])

```

Here you can see the solution curves.

```python
plt.plot(y[:,0])
plt.plot(y[:,1])
```

Below is a nice phase diagram for one of the solutions

```python
fig = plt.figure(figsize=(7,5))
for i in range(1,300,10):
    y = trapezium_vector(xdot,[0,1],0,1,i,2)
    plt.plot(y[:,0],color=np.random.rand(3,1))
```

These equations have the soltuions $te^{-2t}$ and $e^{-t}$. So checking the approximated solution  versus the real solution we get,

```python
err1 = abs(1*e**(-2*1)-y[-1,0])
err2 = abs(e**(-1)-y[-1,1])
print("The error in the first solution is %s"%(err1))
print("The error in the second solution is %s"%(err2))
```

#### Lorentz
A good example of a basic system of differential equations arise in the lorentz system. Here you can see how accurate it is!

```python
def f(t,y):
    s=10
    r=28
    b=8/3
    z = np.matrix(np.zeros((1,3)))
    z[0,0] = s*(y[:,1]-y[:,0])
    z[0,1] = y[:,0]*(r-y[:,2])-y[:,1]
    z[0,2] = y[:,0]*y[:,1]-b*y[:,2]
    return z

z = trapezium_vector(f,[5,5,5],0,50,5000,3)

plt.plot(z[:,0],z[:,2])
```

#### Population Modeling:
Another common application is population modeling!

```python
## f is the fishing constant
f = .6 
def pop(t,y):
    z = np.matrix(np.zeros((1,2)))
    #z[0,0] = -y[:,0]+y[:,0]*y[:,1]
    #z[0,1] = y[:,1]-y[:,0]*y[:,1]
    z[0,0] = -(1+f)*y[:,0]+y[:,0]*y[:,1]
    z[0,1] = (1-f)*y[:,1]-y[:,0]*y[:,1]
    return z
population,time = trapezium_vector(pop,[.5,2],0,100,10000,2,verbose=True)
plt.plot(population[:,0],population[:,1],color='lightseagreen')
plt.ylabel("Number of turtles (in hundreds)")
plt.xlabel("Number of sharks (in hundreds)")
plt.title("Sharks vs. Turtles")
```

```python
fig = plt.figure(figsize=(7,5))
csfont = {'fontname':'cursive'}
plt.plot(time,population[:,0],linestyle="--",color="darkslategrey",label="turtles")
plt.plot(time,population[:,1],color="darkorchid",label="sharks")
legend = plt.legend(loc='upper right', shadow=True, fontsize='large')
plt.title("Sharks vs Sea Turtles",fontsize="x-large",**csfont)
plt.xlabel("Time in years",fontsize="x-large")
plt.ylabel("Population in hundreds",fontsize="x-large")
plt.show()
```

### SICKNESS

Another very common application is modeling sicknesses.

```python
## S - susceptibles (eqn 1), I - infectives (eqn 2), R - removed (eqn 3)
## N- population. Set S(0) = N-I
## a is 1/a part of the year in months! so set a to 4 if the length of infection is 3 months (inverse ) (per year)
## b is how many years immunization lasts -- set to .25 if immunization lasts 4 years (inverse)
## r is how many susceptible encounters (inverse)
a = 4.0
b = 0.25
r = 2.*10**(-3)
N = 1000
def ill(t,y):
    z = np.matrix(np.zeros((1,3)))
    z[0,0] = -r*y[:,1]*y[:,0]+b*y[:,2]
    z[0,1] = r*y[:,1]*y[:,0]-a*y[:,1]
    z[0,2] = a*y[:,1] - b*y[:,2]
    return z
for i in range(1,990,50):
    illness,time = trapezium_vector(ill,[i,1000-i,0],0,7,1000,3,verbose=True)
    plt.plot(illness[:,0],illness[:,1],color=np.random.rand(3,1))
plt.xlabel("Susceptible")
plt.ylabel("Infected")
plt.title("Infection in a population")
```

```python
# now the infection lasts for a year
a = 1
for i in range(1,990,80):
    illness,time = trapezium_vector(ill,[i,1000-i,0],0,7,1000,3,verbose=True)
    plt.plot(illness[:,0],illness[:,1],color=np.random.rand(3,1))
plt.xlabel("Susceptible")
plt.ylabel("Infected")
plt.title("Infection in a population")
```

```python
fig = plt.figure(figsize=(7,5))
for i in range(1,990,80):
    illness,time = trapezium_vector(ill,[i,1000-i,0],0,20,1000,3,verbose=True)
    plt.plot(illness[:,0],illness[:,1],color=np.random.rand(3,1))
plt.ylim(0,500)
plt.xlim(0,600)
plt.xlabel("Susceptible")
plt.ylabel("Infected")
plt.title("Infection in a population")
```

## Celestial Mechanics

The study of celestial mechanics began with Isaac Newton in 1687 with his opus $\it{Principia}$. However, Johannes Kepler was the first to correctly describe the equations of planetary motion. 

His basic formulation,
$$\frac{G*(m1+m2)*(p1-p2)}{dist^3} $$
where the dist is the basic distance formula in 3 dimension and p1/p2 represent the positional coordinates, respectively. 



In order to properly model the planetary motions, I took the masses and coordinates of the outer planets normalized to the mass and position of the sun (where the mass of the sun and inner planets is 1.00000597682 and distances are in AU) and the gravitational constant is $2.95912208286*10^{-4}$. The initial values of the sun are taken to be (0,0,0). All data was taken from Geometric Numerical Integration:Structure-Preserving Algorithms for Ordinary Differential Equations,Hairer E. ,Lubich, C. ,Wanner, G., 2006, Springer-Verlag Berlin Heidelberg.

```python
plt.plot(jupiter_quant[:,0],color='coral')
plt.plot(jupiter_runge_quant[:,0],color='darkseagreen')
plt.ylim(0.000045,0.00006)
plt.ylabel("Total Gravitational Energy")
plt.xlabel("Time in days")
plt.title("Total Energy as a conserved quantity")
```

Here we that even though the midpoint method conserves energy better than the normal euler method it is still not perfect. This is likely due to the fact that they are both only second order (compared to the 4th order RK method).
