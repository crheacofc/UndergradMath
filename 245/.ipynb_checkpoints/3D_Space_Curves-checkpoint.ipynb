```python
def trapstep(func,kappa,tau,t,y,h):
    return y+(func(t,y,tau,kappa)[0]+func(t+h,y+h*func(t,y,tau,kappa)[0],tau,kappa)[0])*h/2, func(t,y,tau,kappa)[1]

def trapezium_SC(func,y_init,kappa,tau,a,b,n):
    y = np.matrix(np.zeros((n,12)))
    quantity = np.matrix(np.zeros((n,4)))
    t_0 = a
    pxg=0;pyg=0;pzg=0 #initial position of curve (gamma)
    Tx=y_init[0];Ty=y_init[1];Tz=y_init[2] ##Initial conditions for TUV
    Nx=y_init[3];Ny=y_init[4];Nz=y_init[5] ##Initial conditions for NUV
    Bx=y_init[6];By=y_init[7];Bz=y_init[8] ##Initial conditions for BUV
    y[0,:]=[pxg,pyg,pzg,Tx,Ty,Tz,Nx,Ny,Nz,Bx,By,Bz]

    #quantity[0,0] = (g*ms)/(((y[0,0])**2+(y[0,2])**2+(y[0,4])**2))**(3/2)
    #quantity[0,1] = m1*y[0,0]*y[0,1]
    #quantity[0,2] = m1*y[0,2]*y[0,3]
    #quantity[0,3] = m1*y[0,4]*y[0,5]
    h = (b-a)/n
    for i in range(1,n):
        t_i = t_0 + i*h
        y[i,:] = trapstep(func,kappa,tau,t_i,y[i-1,:],h)[0]
        #quantity[i,:] = trapstep(func,m1,t_i,y[i-1,:],h)[1]
    return y #, quantity
```

```python
def ydot_3D_SpaceCurve(t,y,tau,kappa):
    z = np.array(np.zeros((1,12)))
    z[0,0:3] = y[:,3:6]
    z[0,3:6] = kappa*y[:,6:9]
    z[0,6:9] = -kappa*y[:,3:6]+tau*y[:,9:12]
    z[0,9:12] = -tau*y[:,6:9]
    
    quant = np.array(np.zeros((1,4)))
    #quant[0,0] = g*ms/dist
    #quant[0,1] = m1*z[0,0]*z[0,1]
    #quant[0,2] = m1*z[0,2]*z[0,3]
    #quant[0,3] = m1*z[0,4]*z[0,5]

    return [z,quant]
```

```python
y = trapezium_SC(ydot_3D_SpaceCurve,[0,1/(np.sqrt(2)),1/(np.sqrt(2)),-1,0,0,0,-1/(np.sqrt(2)),1/(np.sqrt(2))],1/2,1/2,0,10,10000)
```
